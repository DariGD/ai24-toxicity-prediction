NOAEL (No Observed Adverse Effect Level) – это метрика, используемая в токсикологии для обозначения максимальной дозы вещества, при которой не наблюдается никаких неблагоприятных эффектов у организменного объекта (например, человека или лабораторного животного) при проведении исследований. Это значение помогает установить безопасные уровни воздействия различных веществ, таких как химические вещества, лекарства и другие потенциально токсичные агенты.
