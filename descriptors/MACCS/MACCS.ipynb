```python
!pip install rdkit
from rdkit import Chem
from rdkit.Chem import MACCSkeys
def maccs(smile):
  mol = Chem.MolFromSmiles(smile)
  maccs_keys = MACCSkeys.GenMACCSKeys(mol)
  n = maccs_keys.ToBitString()
  p = n[1:]
  return p


```
