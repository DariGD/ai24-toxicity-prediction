## Переводим в единый формат

### Chembl

```python
chembl.head()
```

```python
chembl = chembl.rename({
    'LD_50 (a.u.)': 'LD50 (a.u.)'
}, axis = 1)
```

### Comtox

```python
comptox.head()
```

```python
comptox = comptox.rename({
    'Источник': 'Source',
    'SMILES': 'Smiles',
    'method of administration': 'Method of administration',
    'Metric_1': 'LD50',
    'Metric_1 (LD50)': 'LD50 (a.u.)'
}, axis = 1)

comptox.head()
```

### Toxval

```python
toxval.head()
```

```python
toxval = toxval.rename({
    'source': 'Source',
    'smiles': 'Smiles',
    'method of administration': 'Method of administration',
    'LC_50': 'LC50',
    'LD_50': 'LD50',
    'LC_50 (a.u)': 'LC50 (a.u.)',
    'LD_50 (a.u)': 'LD50 (a.u.)'
}, axis = 1)
```

### Toxric

```python
toxric = toxric.rename({
    'LD_50 (a.u.)': 'LD50 (a.u.)'
}, axis = 1)
```

### Pubchem

```python
pubchem_LD50.head()
```

```python
pubchem_LD50 = pubchem_LD50.rename({
    'smiles': 'Smiles',
    'the experimental animal': 'Exp. Animal',
    'the input method': 'Method of administration',
    'metric_LD50': 'LD50',
    'units': 'LD50 (a.u.)',
}, axis = 1)

pubchem_LD50.head()
```

```python
pubchem_LC50 = pubchem_LC50.rename({
    'smiles': 'Smiles',
    'the experimental animal': 'Exp. Animal',
    'the input method': 'Method of administration',
    'metric_LC50': 'LC50',
    'units': 'LC50 (a.u.)',
}, axis = 1)

pubchem_LC50.head()
```

```python
pubchem = pd.merge(pubchem_LC50, pubchem_LD50, on = ['Smiles', 'Exp. Animal', 'Method of administration'], how = 'outer')
pubchem = pubchem[['Smiles', 'Exp. Animal', 'Method of administration', 'LC50', 'LC50 (a.u.)', 'LD50', 'LD50 (a.u.)']]
pubchem['Source'] = 'PubChem'
pubchem
```

### Union

```python
all_data = pd.concat([chembl, comptox, toxric, toxval, pubchem], axis = 0)
all_data
```

### Немного почистим

#### Exp. Animal

```python
all_data['Exp. Animal'].value_counts()
```

```python
count_animal = all_data['Exp. Animal'].value_counts()

bad_animal = list(count_animal[count_animal < 10].index)
all_data = all_data[all_data['Exp. Animal'].apply(lambda x: x not in bad_animal)]
```

```python
all_data['Exp. Animal'].value_counts()
```

#### Method of administration

```python
all_data['Method of administration'].value_counts()
```

```python
count_input = all_data['Method of administration'].value_counts()

bad_input = list(count_input[count_input < 10].index)
all_data = all_data[all_data['Method of administration'].apply(lambda x: x not in bad_input)]
```

```python
all_data['Method of administration'].value_counts()
```

#### a.u.

```python
[i for i in all_data.columns if '(a.u.)' in i]
```

```python
all_data['LD50 (a.u.)'].value_counts()
```

```python
all_data['LC50 (a.u.)'].value_counts()
```

```python
all_data['LD50 (a.u.)'] = 'mg/kg'
all_data['LC50 (a.u.)'] = 'mg/L'
```
