Приводим способы введения к единообразному виду:

```python
df['Method of administration'] = df['Method of administration'].str.lower()
```

```python
df['Method of administration'] = df['Method of administration'].replace({'не указан': 'not specified', 'dermal' : 'skin',
                                                'ip': 'intraperitoneal', 'i.p.': 'intraperitoneal',
                                                'im': 'intramuscular', 'i.m.': 'intramuscular',
                                                'iv': 'intravenous', 'i.v.': 'intravenous',
                                                'exposure in an artificial soil': 'artificial soil'})
```

```python
df['Method of administration'].unique()
```

Приводим названия животных к единообразному виду:

```python
df['Exp. Animal'] = df['Exp. Animal'].str.lower()
```

```python
df['Exp. Animal'] = df['Exp. Animal'].replace({'mus musculus': 'mouse', 'mus sp.' : 'mouse',
                                                'mice': 'mouse', 'rattus norvegicus': 'rat',
                                                'rats': 'rat', 'sprague-dawley rats': 'rat',
                                                'rabbits': 'rabbit', 'dogs': 'dog'})
```

```python
Построим графики распределения по наиболее часто встречающимся характеристикам в каждом столбце
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_2 = df[['Source', 'LD50', 'LC50', 'Exp. Animal', 'Method of administration', 'Carcinogenicity', 'Hepatoxicity', 'Eye_Irritation', 'Mutagenicity', 'Respiratory_Toxicity', 'NOAEL']]

for column in df_2:
    fig = plt.figure(figsize=(6, 4))
    ax = fig.add_subplot()
    top_values = df[column].value_counts().head(5)
    ax = sns.barplot(x=top_values.index, y=top_values.values)
    ax.set_xlabel(column)
    ax.set_ylabel('Frequency')
    ax.grid()
    
    plt.show()
```

```python
fig , ax = plt.subplots(figsize = (10,10))

sns.heatmap(df_2.corr(numeric_only=True),
            linewidths=0.5, annot=True,cmap='viridis',
            linecolor="white", annot_kws = {'size':12})
```
