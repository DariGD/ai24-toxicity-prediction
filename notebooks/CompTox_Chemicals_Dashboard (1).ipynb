```python
# перевод в SMILES
!pip install pubchempy
import pubchempy as pcp

def cas_to_smiles(cas_number):
    try:
        compound = pcp.get_compounds(cas_number, 'name')
        if compound:
            return compound[0].isomeric_smiles
        else:
            return "SMILES not found"
    except Exception as e:
        return str(e)
```

```python
!pip install rdkit
from rdkit import Chem
from rdkit.Chem import Descriptors

def smiles_to_molecular_weight(smiles):
    molecule = Chem.MolFromSmiles(smiles)
    if molecule is None:
        raise ValueError("Невозможно создать молекулу из SMILES")
    molecular_weight = Descriptors.MolWt(molecule)
    return molecular_weight
```
