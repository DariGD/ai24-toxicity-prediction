```python
import pandas as pd
import numpy as np
```

### Загрузка датасета TOXRIC

Датасет TOXRIC состоит из двух частей, отличающихся способом введения хим. соединения в организм (intravenous/intravenous).
Скачаем обе части датасета из БД TOXRIC

```python
# Примечание! Для того, чтобы скачать файл по ссылке нужно сначла открыть ссылку в браузере или дважды запустить ячейку
LD_50_mouse_0 = pd.read_csv('../data/raw/Acute Toxicity_mouse_intraperitoneal_LD50.csv') # Cпособ введения препарата - intraperitoneal
LD_50_mouse_0.shape
```

```python
LD_50_mouse_0.head()
```

```python
# Примечание! Для того, чтобы скачать файл по ссылке нужно сначла открыть ссылку в браузере или дважды запустить ячейку
LD_50_mouse_1 = pd.read_csv('../data/raw/Acute Toxicity_mouse_intravenous_LD50.csv', delimiter=',') # Cпособ введения препарата - intravenous
LD_50_mouse_1.shape
```

```python
LD_50_mouse_1.head()
```

Теперь объединим оба датасета в один. При этом добавим новый столбец с указанием способа введения хим. соединения в организм (intravenous/intravenous). Пусть для способа intravenous значение будет равно 0, а для ntravenous - 1

```python
# Создадим новый столбец, определяющий метод введения вещества в организм
LD_50_mouse_0['Method of administration'] = 'Intraperitoneal'
LD_50_mouse_1['Method of administration'] = 'Intravenous'
```

```python
LD_50_mouse = pd.concat([LD_50_mouse_0, LD_50_mouse_1], axis=0)
LD_50_mouse.shape
```

```python
LD_50_mouse.info()
```

```python
# Проверим, есть ли в датасете дубли в столбце Canonical SMILES
dublicated = LD_50_mouse[LD_50_mouse['Canonical SMILES'].duplicated()] #Есть 2908 дублей, то есть столько соединений представлено в обоих дафреймах
dublicated
```

```python
# Проверим, есть ли в датасете полные дубли
d = LD_50_mouse.duplicated()
d_rows = LD_50_mouse[d]
print(d.sum()) # видно, что в датасете нет дублей
```

Приведем датафрейм к виду, указанному в issue

|Название поля| Описание|
|---|---|
|Источник|Название исходного источника, откуда это наблюдение|
|Smiles|Уникальный идентификатор химического соединения, подробнее [тут](https://en.wikipedia.org/wiki/Simplified_Molecular_Input_Line_Entry_System)|
|Exp. Animal| Животное, на котором проводился эксперимент|
|method of administration| Способ введения препарата|
|Metric_1| Значение метрики в общих единицах измерениях|
|Metric_1 (a.u)| Единицы измерения метрики 1|
|Metric_2 | ...|
|...|...|

```python
# Удалим ненужные столбцы
LD_50_mouse.drop(['TAID', 'Name', 'IUPAC Name', 'PubChem CID', 'InChIKey'], axis=1, inplace=True)
```

```python
# Теперь создадим новый датафрейм, в который будем добавлять данные из всех датафреймов
ds_toxric = pd.DataFrame(columns=['Source', 'Smiles', 'Exp. Animal', 'Method of administration', 'LD50', 'LD_50 (a.u.)'])
ds_toxric['Smiles'] = LD_50_mouse['Canonical SMILES']
ds_toxric['Source'] = 'TOXRIC'
ds_toxric['Exp. Animal'] = 'Mouse'
ds_toxric['Method of administration'] = LD_50_mouse['Method of administration']
ds_toxric['LD50'] = LD_50_mouse['Toxicity Value']
ds_toxric['LD_50 (a.u.)'] = 'mg/kg'
```

```python
ds_toxric
```
